```python
from datetime import datetime
from pykrx import stock
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
tickers = stock.get_index_ticker_list()
tickers_list = {}
for i in range(len(tickers)):    
    names = stock.get_index_ticker_name(tickers[i])
    tickers_list[names] = tickers[i]
```

```python
tickers_list
```

```python
kospi_tickers = stock.get_market_ticker_list(datetime.today().strftime('%Y%m%d'), market='KOSPI')
kospi_tickers_list = {}
for i in range(len(tickers)):    
    names = stock.get_market_ticker_name(kospi_tickers[i])
    kospi_tickers_list[names] = kospi_tickers[i]
```

```python
class Ecos:
    def __init__(self, auth_key):
        self.auth = auth_key
        self.base_url = 'http://ecos.bok.or.kr/api'
    
    def get_static_table_list(self, static_code=None):
        if static_code is None:
            static_code = ''
        param_list = [
            'StatisticTableList',
            self.auth,
            res_format,
            language,
            start_count,
            end_count,
            static_code
        ]
        param_url = '/'.join(param_list)
        url = f'{self.base_url}/{param_url}'
        res = requests.get(url)
        if res.status_code == 200:
            result = res.json()
            # print(result)
        else:
            print('request failed:', res.status_code)
        df = pd.DataFrame(result['StatisticTableList']['row'])
        return df
    
    def get_static_word(self, word):
        param_list = [
            'StatisticWord',
            self.auth,
            res_format,
            language,
            start_count,
            end_count,
            word
        ]
        param_url = '/'.join(param_list)
        url = f'{self.base_url}/{param_url}'
        res = requests.get(url)
        if res.status_code == 200:
            result = res.json()
            # print(result)
        else:
            print('request failed:', res.status_code)
        df = pd.DataFrame(result['StatisticWord']['row'])
        return df
    
    def get_static_item_list(self, static_code):
        param_list = [
            'StatisticItemList',
            self.auth,
            res_format,
            language,
            start_count,
            end_count,
            static_code
        ]
        param_url = '/'.join(param_list)
        url = f'{self.base_url}/{param_url}'
        res = requests.get(url)
        if res.status_code == 200:
            result = res.json()
            # print(result)
        else:
            print('request failed:', res.status_code)
        df = pd.DataFrame(result['StatisticItemList']['row'])
        return df
    
    def get_static_serch(self, static_code, period, start_time, end_time, item_code1=None, item_code2=None, item_code3=None, item_code4=None):
        if item_code1 is None:
            item_code1 = ''
        if item_code2 is None:
            item_code2 = ''
        if item_code3 is None:
            item_code3 = ''
        if item_code4 is None:
            item_code4 = ''
        
        param_list = [
            'StatisticSearch',
            self.auth,
            res_format,
            language,
            start_count,
            end_count,
            static_code,
            period,
            start_time,
            end_time,
            item_code1,
            item_code2,
            item_code3,
            item_code4
        ]
        param_url = '/'.join(param_list)
        url = f'{self.base_url}/{param_url}'
        res = requests.get(url)
        if res.status_code == 200:
            result = res.json()
            # print(result)
        else:
            print('request failed:', res.status_code)
        df = pd.DataFrame(result['StatisticSearch']['row'])
        return df
    
    def get_key_statistic_list(self):
        param_list = [
            'KeyStatisticList',
            self.auth,
            res_format,
            language,
            start_count,
            '100'
        ]
        param_url = '/'.join(param_list)
        url = f'{self.base_url}/{param_url}'
        res = requests.get(url)
        if res.status_code == 200:
            result = res.json()
            # print(result)
        else:
            print('request failed:', res.status_code)
        df = pd.DataFrame(result['KeyStatisticList']['row'])
        return df
```
